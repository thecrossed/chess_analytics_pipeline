```python
import pandas as pd
import numpy as np
from openpyxl import load_workbook
```

```python
data = pd.read_csv("../../Desktop/rcc_summer_2024/RCC_2024_summer_student_performance_output.csv")
```

```python
data.loc[data['username'] == 'blakey0313']
```

```python
max_class_df = data.groupby(['class_name']).max().reset_index()[['class_name',
                                                 'latest_rating',
                                                 'change']].rename(columns = {"latest_rating":"max_latest_rating",
                                                                             "change":"max_change"})
```

```python
max_ratinng_student = max_class_df[['class_name','max_latest_rating']].merge(data[['class_name',
                         'username',
                         'real_name',
                         'latest_rating']],
                  left_on = ['class_name','max_latest_rating'],
                  right_on = ['class_name','latest_rating'])

max_ratinng_student = max_ratinng_student[['class_name','username','real_name','max_latest_rating']]
max_ratinng_student
```

```python
max_change_student = max_class_df[['class_name','max_change']].merge(data[['class_name',
                         'username',
                         'real_name',
                         'change']],
                  left_on = ['class_name','max_change'],
                  right_on = ['class_name','change'])

max_change_student = max_change_student[['class_name','username','real_name','max_change']]
max_change_student
```

```python
class_avg = data.groupby(['class_name']).mean(numeric_only = True).reset_index()[['class_name',
                                                                     'begin_rating',
                                                                     'latest_rating',
                                                                     'change']]
class_avg
```

```python
max_change_student.to_excel("../../Desktop/rcc_summer_2024/RCC_2024_summer_student_performance_aggr.xlsx", 
                            sheet_name='涨分最高学生', index=False)

max_ratinng_student.to_excel("../../Desktop/rcc_summer_2024/RCC_2024_summer_student_performance_aggr.xlsx", 
                            sheet_name='分数最高学生', index=False)

class_avg.to_excel("../../Desktop/rcc_summer_2024/RCC_2024_summer_student_performance_aggr.xlsx", 
                            sheet_name='班级平均表现', index=False)
```

```python
path = r"../../Desktop/rcc_summer_2024/RCC_2024_summer_student_performance_aggr.xlsx"

book = load_workbook(path)
writer = pd.ExcelWriter(path, engine = 'openpyxl')
writer.book = book

data.to_excel(writer, sheet_name = '学生分数数据')
max_change_student.to_excel(writer, sheet_name = '涨分最高学生')
max_ratinng_student.to_excel(writer, sheet_name = '分数最高学生')
class_avg.to_excel(writer, sheet_name = '班级平均表现')

writer.close()
```
